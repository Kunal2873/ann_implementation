```python
import numpy as np
import pandas as pd
import tensorflow as tf

from tensorflow.keras.models import load_model

import pickle
```

```python
model=load_model('model.h5')
```

```python
with open('ohe.pkl','rb')as file:
    label_encoder_geo=pickle.load(file)
with open('scaler.pkl','rb')as file:
    label_scaler_encoder=pickle.load(file)
with open('label_encoder_gender.pkl','rb')as file:
    label_gender_encoder=pickle.load(file)
```

```python
input_person = {
    "RowNumber": 9999,
    "CustomerId": 15876543,
    "Surname": "Sharma",
    "CreditScore": 720,
    "Geography": "Germany",
    "Gender": "Male",
    "Age": 45,
    "Tenure": 5,
    "Balance": 120000.50,
    "NumOfProducts": 2,
    "HasCrCard": 1,
    "IsActiveMember": 0,
    "EstimatedSalary": 95000.25
}
```

```python
geo_encoded=label_encoder_geo.transform([[input_person['Geography']]]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoded,columns=label_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_person_df=pd.DataFrame([input_person])
input_person_df
```

```python
input_person_df['Gender']=label_gender_encoder.transform(input_person_df['Gender'])
input_person_df
```

```python
# input_person_df=pd.concat([input_person_df.reset_index(drop=True),geo_encoded_df],axis=1)
```

```python
input_person_df
```

```python
input_person_df=pd.concat([input_person_df.drop("Geography",axis=1),geo_encoded_df],axis=1)
input_person_df
```

```python
input_person_df=input_person_df.drop(["RowNumber","CustomerId","Surname"],axis=1)
```

```python
input_person_df
```

```python
#scaling the input data
input_scaled=label_scaler_encoder.transform(input_person_df)
input_scaled
```

```python
prediction=model.predict(input_scaled)
prediction
```

```python
prediction_prob=prediction[0][0]
prediction
```

```python
if prediction_prob>0.5:
    print("yes")
else:
    print("No")
```
